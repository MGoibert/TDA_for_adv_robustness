```python
import pickle
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pandas as pd
```

```python
from tda.rootpath import db_path
from r3d3.experiment_db import ExperimentDB
db_path = '/Users/m.goibert/Documents/Criteo/P2_TDA_Detection/TDA_for_adv_robustness/r3d3.db'

db = ExperimentDB(db_path)
```

```python
experiment_id = 1600962255  # 1600941346
```

```python
df = db.show_experiment(experiment_id)
df
```

```python
def plot(run_id):
    
    try:
        print(df.query(f"run_id=={run_id}")[["attack_type", "attack_backend"]])
    except:
        pass
    
    epsilons = list(sorted(np.linspace(0.0, 0.4, num=11))) if run_id in [0, 1] else [0.0, 1]
    
    for eps in epsilons:
        print(eps)
    
        path = f"../results/{experiment_id}/{run_id}/images_eps_{eps}.pickle"
        images = [x.squeeze().numpy().transpose((1,2,0)) for x in pickle.load(open(path, "rb"))]

        #df = pd.DataFrame(images[0].reshape(-1))
        #df.plot.kde()

        fig=plt.figure(figsize=(20, 20))
        columns = 4
        rows = 1
        for i in range(1, columns*rows +1):
            fig.add_subplot(rows, columns, i)
            plt.imshow(images[i-1])
        plt.show()
```
