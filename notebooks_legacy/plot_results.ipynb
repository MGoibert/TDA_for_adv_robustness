```python
epsilon = list(df_mnist_mlp_fgsm['val'][0].keys())
epsilon = [float(item) for item in epsilon]
epsilon = [epsilon[4]] + epsilon[0:4]
epsilons = {
    "FGSM": epsilon,
    "BIM": epsilon,
    "DeepFool": [0,0.40],
    "CW": [0, 0.40]
}
epsilons
```

```python
mnist_mlp_auc = {
    "FGSM": [list(df_mnist_mlp_fgsm['val'][0].values())[4]]+list(df_mnist_mlp_fgsm['val'][0].values())[0:4],
    "BIM": [list(df_mnist_mlp_bim['val'][0].values())[4]]+list(df_mnist_mlp_bim['val'][0].values())[0:4],
    "DeepFool": [list(df_mnist_mlp_deepfool['val'][0].values())[1]]+[list(df_mnist_mlp_deepfool['val'][0].values())[0]],
    "CW": [list(df_mnist_mlp_cw['val'][0].values())[1]]+[list(df_mnist_mlp_cw['val'][0].values())[0]]
}
mnist_lenet_auc = {
    "FGSM": [list(df_mnist_lenet_fgsm['val'][0].values())[4]]+list(df_mnist_lenet_fgsm['val'][0].values())[0:4],
    "BIM": [list(df_mnist_lenet_bim['val'][0].values())[4]]+list(df_mnist_lenet_bim['val'][0].values())[0:4],
    "DeepFool": [list(df_mnist_lenet_deepfool['val'][0].values())[1]]+[list(df_mnist_lenet_deepfool['val'][0].values())[0]],
    "CW": [list(df_mnist_lenet_cw['val'][0].values())[1]]+[list(df_mnist_lenet_cw['val'][0].values())[0]]
}
svhn_lenet_auc = {
    "FGSM": [list(df_svhn_lenet_fgsm['val'][0].values())[4]]+list(df_svhn_lenet_fgsm['val'][0].values())[0:4],
    "BIM": [list(df_svhn_lenet_bim['val'][0].values())[4]]+list(df_svhn_lenet_bim['val'][0].values())[0:4],
    "DeepFool": [list(df_svhn_lenet_deepfool['val'][0].values())[1]]+[list(df_svhn_lenet_deepfool['val'][0].values())[0]],
    "CW": [list(df_mnist_lenet_cw['val'][0].values())[1]]+[list(df_mnist_lenet_cw['val'][0].values())[0]]
}
svhn_lenet_auc
```

```python
for i in range(4):
    plt.style.use('ggplot')
    plt.plot(epsilons[list(mnist_mlp_auc.keys())[i]],
             mnist_mlp_auc[list(mnist_mlp_auc.keys())[i]], "o:",
             label=list(mnist_mlp_auc.keys())[i], alpha=0.7)
    #plt.plot([0,0.41],[1,1], "--")
    plt.title(f"Non-supervised results for MNIST MLP")
    plt.xlabel(f"L-inf perturbation (epsilon)")
    plt.ylabel(f"AUC")
plt.legend(title="Attack type")
plt.show()
```
